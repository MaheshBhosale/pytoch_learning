```python
import numpy as np
import torch
import matplotlib.pyplot as plt
```

```python
def lin(a, x):
    return np.float64(a*x)
```

```python
def generate_fake_data(a, n):
    x = np.random.uniform(1.0, 10.0, n)
    y = lin(a,x) + np.random.normal(0.0, 2.0,size=n)
    return x, y
```

```python
x, y = generate_fake_data(3,100000)
```

```python
plt.scatter(x=x, y=y, c=['r', 'b'])
```

```python
def mse(ypred, y):
    return ((ypred - y) ** 2).mean()
```

```python
def mse_loss(a, x, y):
    return mse(lin(a, x), y)
```

```python
for a in np.arange(0.0, 6.0, 0.1):
    print("Mean Loss: {} at Weight: {}".format(mse_loss(a, x, y), a))
    plt.plot(a, mse_loss(a, x, y), 'ro') 
```

```python
def gradient_compute(x, y, w):
    return np.float64(2 * x * ( x* w - y))
```

```python
a = np.float64(20.0)
loss = np.float64(0.0)
losses = []
for epoch in range(4):
    for xval,yval in zip(x, y):
        #print(loss , a)
        losses.append(loss)
        a = a  - (0.01 * gradient_compute(xval, yval, a))
        loss = mse_loss(a, xval, yval)
plt.plot(losses)
```
